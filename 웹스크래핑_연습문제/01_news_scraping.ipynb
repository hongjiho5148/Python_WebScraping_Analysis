### 질문1

```python
import requests
from bs4 import BeautifulSoup

req_param = {"sid": "economy"}
url = "https://news.daum.net/{sid}".format(**req_param)
req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36",
    "referer": "https://news.daum.net/",
}

res = requests.get(url, headers=req_header)
res.encoding = "utf-8"
print(url)
print(type(res))
print(res.status_code)

if res.ok:
    soup = BeautifulSoup(res.text, "html.parser")

    a_tags = soup.select("a.item_newsheadline2")
    print(type(a_tags), len(a_tags))  

    for a_tag in a_tags:
         # (a_tag.text를 쓰면 다른 텍스트가 섞일 수 있어 분리)
        title_tag = a_tag.select_one(".tit_txt")
        if title_tag:
            title = title_tag.text.strip()
            link = a_tag["href"]

            print(link)
            print(title)
else:
    # 응답(response)이 error이면 status_code 출력
    print(f'Error Code = {res.status_code}')
```

### 문제2

```python
import requests
from bs4 import BeautifulSoup

section_dict = {
    '기후/환경':'climate', '사회':'society', '경제':'economy', '정치':'politics',
    '국제':'world', '문화':'culture', '생활':'life', 'IT/과학':'tech', '인물':'people'
}

def print_news(section_name):  
    sid = section_dict.get(section_name, 'economy')
    
    url = f'https://news.daum.net/{sid}'
    
    req_header = {
        'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'

    if res.ok:
        print(f"======> {url} {section_name} 뉴스 <======")
        
        soup = BeautifulSoup(res.text, 'html.parser')
        
        a_tags = soup.select("a.item_newsheadline2")
        
        for a_tag in a_tags:

            title_tag = a_tag.select_one(".tit_txt")
            if title_tag:
                title = title_tag.text.strip()
                link = a_tag['href']
                
                print(link)
                print(title)
    else:
        # 응답(response)이 error이면 status_code 출력
        print(f'Error Code = {res.status_code}')

# 함수 실행 예시
print_news('문화')
```
